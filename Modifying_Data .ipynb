```python
import pandas as pd
```

```python
df=pd.read_csv('stackoverflow/survey_results_public.csv',index_col='Respondent')
df.head(5)
```

```python
schema_df=pd.read_csv('stackoverflow/survey_results_schema.csv',index_col='Column')
schema_df.head(5)
```

```python
#change ConvertedComp to SalaryByUSD
df.rename(columns = {'ConvertedComp':'SalaryByUSD'},inplace=True)
df['SalaryByUSD']
```

```python
#change ConvertedComp to SalaryByUSD in schema file
schema_df.rename(index = {'ConvertedComp':'SalaryByUSD'},inplace = True)
schema_df.loc['SalaryByUSD']
```

```python
#change data of two cells
df.loc[[1,2],'SalaryByUSD']='NA'
```

```python
#change the values of many columns for one row
# here we will change the value of age and CompTotal for the first row
df.loc[1,['Age','CompTotal']]=['NA','NA']
df.head(1)
```

```python
#edit data using filter
# edit the data where the column values equal to 'Syrian Arab Republic'
#replace it with 'Syria'
my_filter = (df["Country"] == 'Syrian Arab Republic')
df.loc[my_filter,'Country']='Syria'
my_filter = (df["Country"] == 'Syria')
df.loc[my_filter].head(2)
```

```python
#Change the CurrencySymbol to lowercase
df['CurrencySymbol']=df['CurrencySymbol'].apply(lambda x:str(x).upper())
df['CurrencySymbol']
```

```python
# Change the CurrencySymbol to uppercase
df['CurrencySymbol'] = df['CurrencySymbol'].str.upper()
df['CurrencySymbol']
```

```python
# using applymap() with dataframe, and changing all the data to lowercase.
df.applymap(lambda x:str(x).lower())
```

```python
# using map() with series object
# change value of 'Hobbyist' -----  yes => True, no => False
# we cannot user inplace with map
# so to store the result we assign it
# df['Hobbyist'] = df['Hobbyist'].map({'Yes':'True','No':'False'})
df['Hobbyist'].map({'Yes':'True','No':'False'})
```

```python
#merging two column
df['MajorAndUsedLanguages'] = df['MainBranch'] + '\\' + df['WebframeWorkedWith']
df.head(2)
```

```python
# Delete column
df.drop(columns='MajorAndUsedLanguages',inplace = True)
```

```python
#imporve the data of WebframeWorkedWith data
df['WebframeWorkedWith'] = df['WebframeWorkedWith'].str.replace(';',', ')
df['WebframeWorkedWith']
```
