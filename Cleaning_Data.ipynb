```python
na_values= ['NA','Missing']
df = pd.read_csv('stackoverflow/survey_results_public.csv',index_col = 'Respondent')
len(df)
```

```python
# clean data for the Years Code
df['YearsCode'].head(5)
```

```python
#the type is string so, we have to know what is it
df['YearsCode'].unique()
```

```python
#replace 'Less than 1 year' with zero and 'More than 50 years' with 51
df['YearsCode'].replace('Less than 1 year',0,inplace = True)
df['YearsCode'].replace('More than 50 years',51,inplace = True)
```

```python
df['YearsCode'] = df['YearsCode'].astype(float)
```
