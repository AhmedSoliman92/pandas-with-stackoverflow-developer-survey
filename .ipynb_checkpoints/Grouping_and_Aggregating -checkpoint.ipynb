```python
import pandas as pd
```

```python
df = pd.read_csv('stackoverflow/survey_results_public.csv',index_col='Respondent')
df.head(2)
```

```python
# the median of the values for every column
df.median()
```

```python
# the median of the salary column
df['ConvertedComp'].median()
```

```python
# the number of valid data for every column
df.count()
```

```python
#analyse and explore the data
df.describe()
```

```python
# how many people are Hobbyist
df['Hobbyist'].value_counts()
```

```python
df['PlatformWorkedWith'].value_counts()
```

```python
#What was the primary field of study for the participated programmer?\percentage\
df['UndergradMajor'].value_counts(normalize = True)
```

```python
# first split the object then apply the get_group() function.
Country_group_by = df.groupby(['Country'])
Country_group_by.get_group('Syrian Arab Republic')
```

```python
# display the platforms which are using in developers in every country
Country_group_by['PlatformWorkedWith'].value_counts()
```

```python
# display the platforms which are using in developers in Syria \using filter\
my_filter = df['Country'] == 'Syrian Arab Republic'
df.loc[my_filter]['PlatformWorkedWith'].value_counts()
```

```python
# display the platforms which are using in developers in Syria \using group\
Country_group_by['PlatformWorkedWith'].value_counts().loc['Syrian Arab Republic']
```

```python
# display the percentage of using platforms for every counntry
Country_group_by['PlatformWorkedWith'].value_counts(normalize = True)
```

```python
# the average salary for every country
Country_group_by['ConvertedComp'].median()
```

```python
# the average salary in Syria
Country_group_by['ConvertedComp'].median().loc['Syrian Arab Republic']
```

```python
#Aggregation
# get the median and the mean of salary for the countries
Country_group_by['ConvertedComp'].agg(['median','mean'])
```

```python
# get the median and the mean of salary in Syria
Country_group_by['ConvertedComp'].agg(['median','mean']).loc['Syrian Arab Republic']
```

```python
# how many people know how to use python
Country_group_by['LanguageDesireNextYear'].apply(lambda x : x.str.contains('Python')).sum()
```
