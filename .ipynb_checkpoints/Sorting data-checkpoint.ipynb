```python
import pandas as pd
```

```python
df = pd.read_csv('stackoverflow/survey_results_public.csv', index_col = 'Respondent')
df.head(5)
```

```python
#sorting data acording to Country column (ascending)
df = df.sort_values(by = 'Age', ascending = False)
df.head(5)
```

```python
#sorting data acording to Country column (descending)
df = df.sort_values(by = 'Age', ascending = True)
df.head(5)
```

```python
# sort data according to salary then to age 
df = df.sort_values(by = ['ConvertedComp','Age'],ascending = True).head(5)
```

```python
# restore the default sort
df.sort_index()
```

```python
# display values of  sorted column
df['Country'].sort_values()
```
